Feature Engineering Using The Percentile Method To Remove Outliers

```python
#An Outlier is an unusual data point that is very different from the others.
#In Other words they are assumed to be incorrect data points.
# Let's take a Good example of having a dataset saying a Human of 1000 years, Is an outlier Because it is Impossible.
```

                                         #Simple Data Set To Begin With

#In the Work Below, We are Going To be dealing with a simple Data set of height To remove the Outliers

```python
import pandas as pd
```

```python
df = pd.read_csv('height.csv')
df
```

```python
#Checking the number of Rows and Columns
df.shape
```

```python
#Checking brief description of the dataset
df.describe()
#As You see below, It will give Us the Mean, Minimum value, max value, and Many others
```

```python
#Here we are trying to initialize that the maximum height is 95% of the heights.
#U will need to have a good understanding of Percentiles to get this
max_height_should_be =df['height'].quantile(0.95)
max_height_should_be
```

```python
#Displaying The outlier which is greater than the maximum height
df[df['height']>max_height_should_be]
```

```python
#We are trying to say That the minimum Height should be 5% of the height data
min_height_should_be = df['height'].quantile(0.05)
min_height_should_be
```

```python
#Displaying The outlier which is less than the minimum height
df[df['height']<min_height_should_be]
```

```python
#Removing The Outliers and Taking The data we need
df[(df['height']<max_height_should_be) & (df['height']>min_height_should_be)]
```

```python
#We have now removed the outliers From Our Dataset and Therefore.
#We are ready to go Using the Dataset.
#Since Feature Engineering Is done.
```

```python
#We shall take the dataset as Data Frame 2
df2 = pd.read_csv('BHP.csv')
df2
```

```python
#Checking the number of Rows and Columns
df2.shape
```

```python
#Checking brief description of the dataset
df2.describe()
#As You see below, It will give Us the Mean, Minimum value, max value, and Many others
```

```python
min_price,max_price = df2.price.quantile([0.001,0.999])
min_price,max_price
```

```python
#getting The outlier price below the Minimum Price
df2[df2.price < min_price]
```

```python
#getting The outlier price above the maximum Price
df2[df2.price > max_price]
```

```python
#Now removing The Outliers and Remaining With The Correct Data Fpor consumption
#We can go ahead and Create another variable as df3 for the final Clean Dataset
df3 = df2[(df2.price> min_price)&(df2.price<max_price)]
df3
```
